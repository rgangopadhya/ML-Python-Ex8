```python
movie_list = loadMovieList(root_in + '/movie_ids.txt')
```

```python
#now input a users ratings
my_ratings = np.zeros((len(movie_list), 1))
my_ratings[0] = 4
my_ratings[97] = 2

# We have selected a few movies we liked / did not like and the ratings we
# gave are as follows:
my_ratings[6] = 3
my_ratings[11]= 5
my_ratings[53] = 4
my_ratings[63]= 5
my_ratings[65]= 3
my_ratings[68] = 5
my_ratings[182] = 4
my_ratings[225] = 5
my_ratings[354]= 5
for i, rate in enumerate(my_ratings):
    if rate>0:
        print "Rated {0} for {1}".format(str(int(rate)), movie_list[i])
```

```python
#pd contains the loaded ratings
Y = np.hstack((my_ratings, pd['Y']))
my_r = np.zeros(my_ratings.shape)
my_r[my_ratings != 0] = 1
R = np.hstack((my_r, pd['R']))
```

```python
#normalize the ratings, and initialize X and theta
(Ynorm, Ymean) = normalizeRatings(Y, R)
num_users = Y.shape[1]
num_movies = Y.shape[0]
num_features = 10
X = np.random.normal(size = (num_movies, num_features))
Theta = np.random.normal(size = (num_users, num_features))
initial_param = np.vstack((X, Theta)).flatten()
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        
lambd = 10
theta = optimize.minimize(lambda t: cofiCostFunc(t, Ynorm, R, num_users, num_movies, num_features, lambd), 
   initial_param, method = 'CG', jac=True, options = {"maxiter" : 200})
```

```python
theta
```

```python
#make recommendations based on learned parameters
params = theta['x']
X = np.reshape(params[:num_movies*num_features], (num_movies, num_features))
Theta = np.reshape(params[num_movies*num_features:], (num_users, num_features))

p = np.dot(X, np.transpose(Theta))
my_predictions = p[:, 0] + Ymean.flatten()
#pick top 10 predicted
sort_indices = np.argsort(my_predictions)[::-1]
for j in xrange(10):
    i = sort_indices[j]
    print "Predicted rating {0:0.2f} for movie {1}".format(my_predictions[i], movie_list[i])
#Original ratings provided 
for i, rate in enumerate(my_ratings):
    if rate>0:
        print "Rated {0} for {1}".format(str(int(rate)), movie_list[i])
```
